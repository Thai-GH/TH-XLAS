```python
#1
import imageio.v2 as iio
import numpy as np

img = iio.imread('bird.png')

# Tách các kênh màu
r = img.copy()
r[:, :, 1] = 0
r[:, :, 2] = 0
iio.imwrite('bird_red.png', r)

g = img.copy()
g[:, :, 0] = 0
g[:, :, 2] = 0
iio.imwrite('bird_green.png', g)

b = img.copy()
b[:, :, 0] = 0
b[:, :, 1] = 0
iio.imwrite('bird_blue.png', b)
```

```python
#2
import imageio.v2 as iio
import numpy as np

img = iio.imread('bird.png')

# Đổi RGB → BGR
bgr = img[:, :, ::-1]
iio.imwrite('bird_bgr.png', bgr)

# RGB → GRB
grb = img[:, :, [1, 0, 2]]
iio.imwrite('bird_grb.png', grb)

# RGB → BRG
brg = img[:, :, [2, 0, 1]]
iio.imwrite('bird_brg.png', brg)
```

```python
#3
import imageio.v2 as iio
import numpy as np
import colorsys
import matplotlib.pyplot as plt

img = iio.imread('bird.png') / 255.0  # Chuẩn hóa

# Vector hóa hàm RGB to HSV
rgb2hsv = np.vectorize(colorsys.rgb_to_hsv)
h, s, v = rgb2hsv(img[:, :, 0], img[:, :, 1], img[:, :, 2])

# Nhân lại để hiển thị đúng thang độ
iio.imwrite('hue.png', (h * 255).astype(np.uint8))
iio.imwrite('saturation.png', (s * 255).astype(np.uint8))
iio.imwrite('value.png', (v * 255).astype(np.uint8))
```

```python
#4
import imageio.v2 as iio
import numpy as np
import colorsys

img = iio.imread('bird.png') / 255.0
rgb2hsv = np.vectorize(colorsys.rgb_to_hsv)
h, s, v = rgb2hsv(img[:, :, 0], img[:, :, 1], img[:, :, 2])

# Chỉnh H, V
h_new = h / 3.0
v_new = v * 0.75

hsv2rgb = np.vectorize(colorsys.hsv_to_rgb)
r_new, g_new, b_new = hsv2rgb(h_new, s, v_new)

# Kết hợp lại thành ảnh RGB
rgb_new = np.stack((r_new, g_new, b_new), axis=2)
iio.imwrite('adjusted_hsv_rgb.png', (rgb_new * 255).astype(np.uint8))
```

```python
#6
import os
import numpy as np
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pyplot as plt

input_dir = 'Exercise'
output_dir = 'Filtered'
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith('.png') or file.endswith('.jpg'):
        path = os.path.join(input_dir, file)
        img = iio.imread(path, mode='F').astype(np.uint8)

        mean = sn.convolve(img, np.ones((5, 5)) / 25).astype(np.uint8)
        median = sn.median_filter(img, size=5)
        maxf = sn.maximum_filter(img, size=5)

        iio.imwrite(os.path.join(output_dir, f'{file}_mean.png'), mean)
        iio.imwrite(os.path.join(output_dir, f'{file}_median.png'), median)
        iio.imwrite(os.path.join(output_dir, f'{file}_max.png'), maxf)

print("Đã lọc xong. Hãy so sánh ảnh kết quả để đánh giá filter nào tốt nhất.")
```

```python
#7
from skimage import filters

edge_dir = 'Edges'
os.makedirs(edge_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith('.png') or file.endswith('.jpg'):
        path = os.path.join(input_dir, file)
        img = iio.imread(path, mode='F').astype(np.uint8)

        # Khử nhiễu trước
        clean = sn.median_filter(img, size=5)

        # Phát hiện biên
        edges = filters.sobel(clean)
        edges_uint8 = (edges * 255).astype(np.uint8)

        iio.imwrite(os.path.join(edge_dir, f'{file}_sobel.png'), edges_uint8)
```

```python
#8
import random

rgb_random_dir = 'RGB_Shuffle'
os.makedirs(rgb_random_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith('.png') or file.endswith('.jpg'):
        path = os.path.join(input_dir, file)
        img = iio.imread(path)
        
        # Khử nhiễu
        img = sn.median_filter(img, size=(1, 1, 1))  # không làm thay đổi ảnh màu
        
        # Đổi vị trí kênh màu
        channels = [0, 1, 2]
        random.shuffle(channels)
        img_shuffled = img[:, :, channels]

        iio.imwrite(os.path.join(rgb_random_dir, f'{file}_rgb_shuffle.png'), img_shuffled)
```

```python
#9
import colorsys

hsv_random_dir = 'HSV_Shuffle'
os.makedirs(hsv_random_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith('.png') or file.endswith('.jpg'):
        path = os.path.join(input_dir, file)
        img = iio.imread(path)
        img = sn.median_filter(img, size=(1, 1, 1))

        # Chuyển đổi từng pixel sang HSV và đảo trộn
        hsv = np.zeros_like(img, dtype=float)
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                r, g, b = img[i, j] / 255.0
                h, s, v = colorsys.rgb_to_hsv(r, g, b)
                hsv[i, j] = [h, s, v]

        # Xử lý ngẫu nhiên không trùng: nhân H, S, V với hệ số khác nhau
        hsv[:, :, 0] = (hsv[:, :, 0] + random.uniform(0.1, 0.5)) % 1.0
        hsv[:, :, 2] = hsv[:, :, 2] * random.uniform(0.5, 1.0)

        # Chuyển lại RGB
        rgb_new = np.zeros_like(img)
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                h, s, v = hsv[i, j]
                r, g, b = colorsys.hsv_to_rgb(h, s, v)
                rgb_new[i, j] = [int(r * 255), int(g * 255), int(b * 255)]

        iio.imwrite(os.path.join(hsv_random_dir, f'{file}_hsv_random.png'), rgb_new.astype(np.uint8))
```
